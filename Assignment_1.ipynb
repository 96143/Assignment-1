```
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
 
import warnings
warnings.filterwarnings("ignore")
 
x1 = np.random.normal(0, 1, 100000)
x2 = np.random.normal(0, 1, 100000)
 
V = np.power(x1,2) + np.power(x2,2)
 
sns.distplot(V, hist_kws={'cumulative': True} , kde_kws={'cumulative': True}, hist = False )
plt.title('CDF OF V')
plt.show()
plt.clf()
 
 
sns.distplot(V, hist = False )
plt.title('PDF OF V')
plt.show()
plt.clf()

```

```
x = np.linspace(-2, 25, 100) 
y = np.exp(-x/2)
  
fig = plt.figure(figsize = (10, 5)) 
# Create the plot 
plt.plot(x, y) 
plt.title('PDF OF V')  
# Show the plot 
plt.show() 
```
